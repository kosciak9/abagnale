```python
import math
from functools import reduce
from collections import Counter

from tqdm.auto import tqdm
import plotly.graph_objects as go

from scraper import Web, Site
```

```python
web = Web.from_zip('./outputs/iosco.gz')
web
```

```python
entities = {
    site: site.entities
    for site in tqdm(web)
}
len(entities)
```

```python
list(entities.values())[:2]
```

```python
counter = Counter(reduce(
    lambda current, ents: current + list(ents),
    entities.values(),
    list()
))
counter.most_common(10)
```

```python
len(counter)
```

```python
go.Figure(
    layout=dict(
        title='Distribution of entity frequencies',
        xaxis_title='Log number of times an entity appears',
        yaxis_title='Number of entities'
    ),
    data=[
        go.Histogram(
            x=[math.log(count) for count in counter.values()]
        )
    ]
)
```
