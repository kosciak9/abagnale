```python
import json
from tqdm.auto import tqdm
```

```python
with open('../data/outputs/web.json') as file:
    web = json.loads(file.read())
web[0]
```

```python
with open('../data/outputs/entities.json') as file:
    entities = json.loads(file.read())
entities['www.10cryptomarket.com']
```

```python
found_sites = [site for site in tqdm(web) if 'bitcoin' in site['text'].lower() and site['lang'] == 'pl']
[site['url'] for site in found_sites]
```

```python
found_entities = {
    entity: data
    for entity, data in entities.items()
    if set(data['source_urls']) & set(site['url'] for site in found_sites)
}
found_entities
```
