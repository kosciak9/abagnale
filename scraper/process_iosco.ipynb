```python
import pandas as pd
from bs4 import BeautifulSoup
```

```python
with open('./sources/iosco.html', 'r') as iosco_source:
    html = iosco_source.read()
html[:20]
```

```python
soup = BeautifulSoup(html, 'lxml')
table = soup.find_all('table')[0]

rows = [
    row
    for row in [
        row.find_all('td')
        for row in table.find_all('tr')
    ]
    if len(row) > 0
]
data = pd.DataFrame.from_dict(dict(
    company_name=[
        next(row[0].children).strip()
        for row in rows
    ],
    jurisdiction=[
        next(row[2].children).strip()
        for row in rows
    ],
    report_date=[
        next(row[3].find_all('a', href=True)[0].children)
        for row in rows
    ],
    report_url=[
        row[3].find_all('a', href=True)[0]['href']
        for row in rows
    ],
))
data
```

```python
data.to_csv('./outputs/iosco.csv')
```
